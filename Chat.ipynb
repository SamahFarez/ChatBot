```python
#Tokenization
text = corpus 
sentence_list = nltk.sent_tokenize(text) # a list of sentences
```

```python
#Print list
print(sentence_list)
```

```python
#Function to return random greet response to user
def greet(text) :
  text=text.lower()

  #bots greet
  bot_greet = ['howdy', 'sup' , 'wassup', 'hi', 'hiii', 'hey', 'hello', 'slm', 'hola' , 'welcome', 'Gm' ]
  #user greet
  user_greet = ['hi', 'hola' , 'sup', 'hello', 'greetings', 'wassup']


  for word in text.split():
    if word in user_greet:
      return random.choice(bot_greet)
```

```python
def index_sort(list_var):
  length = len(list_var)
  list_index = list( range(0, length))

  x = list_var
  for i in range(length):
    for j in range(length) :
      if x[list_index[i]] > x[list_index[j]] :
      # Swap
        temp = list_index[i]
        list_index[i] = list_index[j]
        list_index[j]= temp

  return list_index
```

```python
#Bot response
def bot_respo(user_input) :
  user_input = user_input.lower()
  sentence_list.append(user_input)
  bot_respo = ''
  cm = CountVectorizer().fit_transform(sentence_list)
  similarity_scores = cosine_similarity(cm[-1], cm)
  similarity_scores_list = similarity_scores.flatten()
  index = index_sort(similarity_scores_list)
  index = index[1:]
  response_flag = 0
  
  j = 0
  for i in range(len(index)):
    if similarity_scores_list[index[i]] > 0.0 :
      bot_respo = bot_respo + ' ' + sentence_list[index[i]]
      response_flag = 1
      j = j + 1
    
    if j > 2:
      break;

    if response_flag == 0:
      bot_respo = bot_respo + ' ' + "Beep beep boop beep ??"

    sentence_list.remove(user_input)

    return bot_respo
```

```python
#Start chat
print( ' Bot: i am ur bot, i will answer ur questions, to exit type bye ')

exit_list = [ 'cya' , 'exit' , 'see ya', 'bye' , 'quit' , 'good night', 'later', 'talk to u later', 'goodbye', 'adios', 'ciao', ]
while(True):
  user_input = input()
  if user_input.lower() in exit_list:
    print(' Chat with you later ;)')
    break
  else:
    if greet(user_input) != None:
      print('Bot: ' + greet(user_input))
    else:
      print('Bot: ' + bot_respo(user_input))
```
